### INTRODUCTION

In my project I will work with the dataset named **TIP DATASET**. This dataset describe what one waiter recorded information about each tip he received over a period of a few months working in one restaurant. He collected several data with eight variables in a dataframe compound 244 observations.

The variables created by waiter were:
    * TipPercentage - a numeric vector, the tip written as a percentage (0-100) of the total bill.
    * Bill - a numeric vector, the bill amount (dollars).
    * Tip - a numeric vector, the tip amount (dollars).
    * Gender - a factor with levels Female Male, gender of the payer of the bill.
    * Smoker - a factor with levels No Yes, whether the party included smokers.
    * Weekday - a factor with levels Friday, Saturday, Sunday, Thurday, day of the week.
    * Time – a fator with levels Day Night, rough time of day.
    * PartySize – a numeric vector, numbers of people in party.

### REGRESSION

In regression analysis, you usually consider some phenomenon of interest and have a number of observations. Each observation has two or more features. Following the assumption that (at least) one of the features depends on the others, you try to establish a relation among them. The dependent features are called the **dependent variables, outputs, or responses.** The independent features are called the **independent variables, inputs, or predictors.**

#### Regression Line

A regression line is a straight line that describes how a response variable y changes as an explanatory variable x changes. We often use a regression line to predict the value of y for a given value of x.

##### "Reg Plot" and "LM Plot" are two important types of figure that we can plot using Seaborn. They have similar appearance, but have different functionality.

#### "LM Plot"

Has three mandatory parameters. These are *X-axis, Y-axis* and the *dataset*. The rest the parameters are optional and we may use as per project require. The parameter called **hue** is an optional parameter but wildely used. **Hue** parameter split the data by colunms and help to group it.

```python
sns.set(color_codes=True)
sns.lmplot(x = "total_bill", y = "tip", data=Tip_data);  # function that is used to visualize a linear relationship as determined through regression.
```

The **Linear Line** is the best trend of the *tip* gave by customers in accordance of the *total bill*. The *points* are tips that the customers gave and the *points* top right that are far away from this line are called **outliers** in the dataset. The algorithm predict the best fit for understanding the trend in behavior of visiting customers. This shadow is actually the **statistical mean** that means the prediction of tip value in this restaurant on a daily basis. We can use all information to better understand the data adding more features to the plot and for this we will use **Hue** parameter. 

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=Tip_data); # function divide the data per column and show the variables.
```

This plot above show us if a *Smoker* in general *tip* **more** or **less** *Non-Smoker*.

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=Tip_data, markers=["o", "x"], palette="Set2");
```

This plot is the same as printed above but here we create different makers to describe the data. It is very interesting to use and we can remove the legend as you decide.

#### "Reg Plot"

Performs a simple linear regression model fit and plot. The difference between regplot and lmplot are: 
        
        ° The shape of plot 
        ° Has mandatory input parameter flexibility, it means X and Y do not necessarily require strings as input. We can use other formats such as NumPy arrays, Pandas Series or Pandas DataFrame.

```python
sns.regplot(x="total_bill", y="tip", data=Tip_data, color="m" ); # function that is used to visualize a linear relationship as determined through regression.
```

If you compare the *lmplot* plot and *regplot* plot you will see that the resulting are identical differentiating only the shape.

#### Controlling the size and shape of the plot.

```python
f, ax = plt.subplots(figsize=(5, 6))
sns.regplot(x="total_bill", y="tip", data=Tip_data, ax=ax);
```

***regplot*** is an “axes-level” function draws onto a specific axes. This means that you can make multi-panel figures yourself and control exactly where the regression plot goes. 

```python
sns.lmplot(x="total_bill", y="tip", col="day", data=Tip_data, col_wrap=2, height=3);
```

***lmplot*** is controlled through the *FacetGrid* interface using the size and aspect parameters, which apply to each facet in the plot, not to the overall figure itself.

More example of ***lmplot***:

```python
sns.lmplot(x="total_bill", y="tip", col="day", data=Tip_data, aspect=.5);
```

#### Plotting a regression in other contexts.

More complex plot that uses *jointplot* with *regplot* to show the linear regression.

```python
sns.jointplot(x="total_bill", y="tip", data=Tip_data, kind="reg");
```

Using the *pairplot* function with kind="reg" combines *regplot* and *PairGrid* to show the linear relationship between variables in a dataset. Take care to note how this is different from *lmplot*. In the figure below, the two axes do not show the same relationship conditioned on two levels of a third variable; rather, *PairGrid* is used to show multiple relationships between different pairings of the variables in a dataset:

```python
sns.pairplot(Tip_data, x_vars=["total_bill", "size"], y_vars=["tip"], height=5, aspect=.9, kind="reg");
```

Like *lmplot*, but unlike *jointplot*, conditioning on an additional categorical variable is built into *pairplot* using the hue parameter:

```python
sns.pairplot(Tip_data, x_vars=["total_bill", "size"], y_vars=["tip"], hue="smoker", height=8, aspect=.6, kind="reg");
```
