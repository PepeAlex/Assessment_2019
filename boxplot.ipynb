```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Tip_data = pd.read_csv("https://github.com/mwaskom/seaborn-data/raw/master/tips.csv")
```

```python
# boxplot Tip vs Gender
sns.boxplot(x="tip", y="sex", data=Tip_data)
plt.show()

# boxplot Tip vs Day
sns.boxplot(x="tip", y="day", data=Tip_data)
plt.show()

# boxplot Tip vs Time
sns.boxplot(x="tip", y="time", data=Tip_data)
plt.show()

# boxplot Tip vs Smoker
sns.boxplot(x="tip", y="smoker", data=Tip_data)
plt.show()
```
