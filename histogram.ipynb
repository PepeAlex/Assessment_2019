```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Tip_data = pd.read_csv("https://github.com/mwaskom/seaborn-data/raw/master/tips.csv")
```

```python
# set the backround colur of the histogram graph
sns.set(style="whitegrid")

# set the bar colour, figure size and bin size of the graphs
Tip_data.hist(color = "green", figsize = (12, 8), bins = 10)
# plot the graph
plt.show()
```
